### Programmation Orientée Objet (POO) en Python

La programmation orientée objet (POO) permet de créer des entités (objets) que l'on peut manipuler.

Une classe regroupe des fonctions et des attributs qui définissent un objet. On appelle par ailleurs les fonctions d'une classe des " méthodes ".

Créons une classe Voiture avec 2 attributs, 4 méthodes et 1 constructeur :

```python
class Voiture:
  "Définition d'une voiture"
  # Attributs
  couleur = ""
  immatriculation = ""
  # Méthodes
  def demarrer(self):
    self.demarree = True
    print("Moteur démarré")
    return
  def accelerer(self):
    return
  def freiner(self):
    return
  def arreter(self):
    return
  # Constructeur
  def __init__(self):
    self.demarree = False
    print("Création d'un nouvel objet instance de la classe Voiture")
```

###Méthodes de classe

```python
class Moto:
    nb = 0

    def __init__(self):
        print("Création d'un objet de type moto")
        Moto.nb += 1
        print("Il y en a maintenant ", Moto.nb)

    @classmethod
    def get_nb(cls):
        return Moto.nb
```

```python
print("Démarrage : Nombre de motos = ", Moto.get_nb())
moto1 = Moto()
print("Nombre de motos = ", Moto.get_nb())
moto2 = Moto()
print("Nombre de motos = ", Moto.get_nb())
```
